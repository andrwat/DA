```python
hsim = pd.read_csv(DATA_DIR + 'hsi_monthly.csv')
hsim.describe()
```

```python
hsim
```

```python
hsim.drop(hsim.columns[[7,8,9,10]],1,inplace=True)
```

```python
hsim
```

```python
hsim['Range']=hsim['Close'] - hsim['Open']
```

```python
hsim['Pct'] = hsim['Range']/hsim['Open']
```

```python
hsimup = hsim[hsim.Pct >= 0].sort('Pct')
```

```python
hsimdown = hsim[hsim.Pct < 0].sort('Pct')
```
